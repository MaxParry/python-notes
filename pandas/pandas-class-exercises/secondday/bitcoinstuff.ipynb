Read in both of the CSV files and print out their DataFrames

Perform an inner merge that combines both DataFrames on the "Date" column

Rename the columns within the newly merged DataFrame so that the headers are more descriptive

Create a summary table that includes the following information: Best Bitcoin Open, Best Dash Open, Best Bitcoin Close, Best Dash Close, Total Bitcoin Volume, Total Dash Volume

Total Bitcoin Volume and Total Dash Volume should be calculated to have units of "millions" and be rounded to two decimal places


```python
import pandas as pd
```

```python
csv1 = pd.read_csv('/Users/maxparry/GitRepos/python-notes/pandas/input_data/bitcoin.csv')
csv2 = pd.read_csv('/Users/maxparry/GitRepos/python-notes/pandas/input_data/dashprice.csv')
```

```python
bit = pd.DataFrame(csv1)
dsh = pd.DataFrame(csv2)
```

```python
bit.head()
```

```python
dsh.head()
```

```python
inner = pd.merge(bit, dsh, on=['Date'], how='inner')
```

```python
inner.head()
```

Rename the columns within the newly merged DataFrame so that the headers are more descriptive

```python
newinner = inner.rename(columns={'Open_x':'Bitcoin Open',
                                 'High_x':'Bitcoin High',
                                 'Low_x':'Bitcoin Low',
                                 'Close_x':'Bitcoin Close',
                                 'Volume_x':'Bitcoin Volume',
                                 'Market Cap_x':'Bitcoin Market Cap',
                                 'Open_y':'Dash Open',
                                 'High_y':'Dash High',
                                 'Low_y':'Dash Low',
                                 'Close_y':'Dash Close',
                                 'Volume_y':'Dash Volume',
                                 'Market Cap_y':'Dash Market Cap'})
```

```python
newinner.head()
```

Create a summary table that includes the following information: Best Bitcoin Open, Best Dash Open, Best Bitcoin Close, Best Dash Close, Total Bitcoin Volume, Total Dash Volume

```python
summary = pd.DataFrame({'Bitcoin':[(newinner['Bitcoin Open'].max()),
                                  (newinner['Bitcoin Close'].max()),
                                  (str((newinner['Bitcoin Volume'].sum() / 1000000).round(decimals=2)) + ' millions')],
                        'Dash':[(newinner['Dash Open'].max()),
                                (newinner['Dash Close'].max()),
                                (str((newinner['Dash Volume'].sum() / 1000000).round(decimals=2)) + ' millions')]},
                       index=['Best Open', 'Best Close', 'Total Volume'])
```

```python
summary
```
